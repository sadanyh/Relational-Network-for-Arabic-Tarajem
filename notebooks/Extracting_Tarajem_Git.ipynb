# A function for extracting the attributes from the biographical data

```python
def extract(html_f):
    with open(html_f) as fp:
        
        
        soup = BeautifulSoup(fp, 'html.parser')
        
        try:
            results = soup.find('div', class_="toggle-content")
            summary = results.p.text.replace("\n", '')
        except Exception:
            summary = None
        try:
            title = soup.find("div", class_="col-md-10")
            name = title.h1.get_text()
        except Exception:
            name = None
        try:
            category = soup.find_all('span', class_="size-13 text-muted pull-right label label-primary")
            dob = category[0].get_text()
            
        except Exception:
            dob = None
        try:
            category = soup.find_all('span', class_="size-13 text-muted pull-right label label-primary")
            gender = category[1].get_text()
        except Exception:
            gender = None
        try:
            death = soup.find ('span', class_="size-13 text-muted pull-right label label-warning")
            dod = death.get_text()
        except Exception:
            dod = None
        try:
            age = soup.find ('span', class_="size-13 text-muted pull-right label label-brown")
            era = age.get_text()
        except Exception:
            era = None    
        diction = {"summary":summary, 
                 "name": name,
                 "date_of_birth":dob,
                 "date_of_death":dod,
                 "gender": gender,
                 "age": era}
        
        return diction
    
    
```

# Separating the Gender attribute

```python
# moving gender from date of birth to gender
import numpy as np
cond1=clean_df.date_of_birth.str.contains('رجال')
cond2=clean_df.date_of_birth.str.contains('نساء')
cond3=clean_df.date_of_birth.str.contains('كنى رجال')
cond4=clean_df.date_of_birth.str.contains('كنى نساء')
cond5=clean_df.date_of_birth.str.contains('مجاهيل رجال')
clean_df['gender']=np.where(cond1,clean_df.date_of_birth,clean_df.gender)
```

```python
# where cond is false (i.e. no رجال is in gender replace the value in gender with that of in date_of_birth)

clean_df['gender']=np.where(cond2,clean_df.date_of_birth,clean_df.gender) 
clean_df['gender']=np.where(cond3,clean_df.date_of_birth,clean_df.gender)
clean_df['gender']=np.where(cond4,clean_df.date_of_birth,clean_df.gender)
clean_df['gender']=np.where(cond5,clean_df.date_of_birth,clean_df.gender)
```

```python
# creating one text file for all the data

def data_to_text(html_f):
    with open(html_f) as fp:
        
        
        soup = BeautifulSoup(fp, 'html.parser')
        results = soup.find('div', class_="toggle-content")
        try:
            summary = results.p.text
        except Exception:
            summary = ' empty '
        try:
            title = soup.find("div", class_="col-md-10")
            name = title.h1.get_text()
        except Exception:
            name = ' empty '
        try:
            birth = soup.find ('span', class_="size-13 text-muted pull-right label label-light")
            dob = birth.get_text()
        except Exception:
            dob = ' empty '
        try:
            death = soup.find ('span', class_="size-13 text-muted pull-right label label-warning")
            dod = death.get_text()
        except Exception:
            dod = ' empty '
        try:
            age = soup.find ('span', class_="size-13 text-muted pull-right label label-brown")
            era = age.get_text()
        except Exception:
            era = ' empty '   
        text_data = summary+" "+ name+" "+dob+" "+dod+" "+era
                 
        
        return text_data
```
